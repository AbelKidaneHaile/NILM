Outlier Detection

```python
feature_df = df[FEATURE_COLS]
```

```python
z_scores = (feature_df - feature_df.mean()) / feature_df.std()
```

```python
threshold = 3
outliers = (np.abs(z_scores) > threshold)
```

```python
outlier_rows = df[outliers.any(axis=1)]
print(f"Detected {len(outlier_rows)} outliers using Z-score")
```

```python
Q1 = feature_df.quantile(0.25)
Q3 = feature_df.quantile(0.75)
IQR = Q3 - Q1
outliers = (feature_df < (Q1 - 1.5 * IQR)) | (feature_df > (Q3 + 1.5 * IQR))
outlier_rows = df[outliers.any(axis=1)]
outlier_rows
```
